## Step 3: Verify GPU & Config

```python
import torch
print(f"GPU Available: {torch.cuda.is_available()}")
if torch.cuda.is_available():
    print(f"GPU Name: {torch.cuda.get_device_name(0)}")
    print(f"GPU Memory: {torch.cuda.get_device_properties(0).total_memory / 1e9:.1f} GB")
```

## Step 4: Update Config for Colab

```python
import yaml

# Update config to use GPU and larger batch size
config = {
    'model_name': 'distilbert-base-uncased',
    'task': 'sentiment',
    'max_seq_length': 128,
    'batch_size': 32,  # Larger batch on GPU
    'learning_rate': 0.00002,
    'num_epochs': 3,
    'warmup_steps': 100,
    'log_every': 50,
    'eval_every': 200,
    'checkpoint_dir': './runs',
    'seed': 42,
    'device': 'cuda' if torch.cuda.is_available() else 'cpu'
}

with open('config_colab.yaml', 'w') as f:
    yaml.dump(config, f)

print("Config updated:")
print(yaml.dump(config))
```

## Step 6: Download Results

```python
import os
import shutil
from google.colab import files

# Create zip of checkpoints
if os.path.exists('runs'):
    shutil.make_archive('toy_finetune_results', 'zip', 'runs')
    files.download('toy_finetune_results.zip')
    print("✓ Downloaded results to your machine")
else:
    print("⚠ No checkpoints found. Check training output above.")
```
